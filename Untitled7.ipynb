```python
import pandas as pd
import csv
```

```python
df = pd.read_csv("bi.csv", encoding = "latin1")
df.head()
```

```python
#Part 1 - Data Cleaning
#1. Check dataset structure
# Display the shape of the dataset
print("Shape of DataFrame:", df.shape)

# Display information about columns and data types
print("\nInfo about DataFrame:")
print(df.info())

# Display the first 5 rows
print("\nFirst 5 rows:")
print(df.head())

# Separate columns by type
categorical_cols = ['fNAME', 'lNAME', 'gender', 'country', 'residence', 'prevEducation']
numerical_cols = ['Age', 'entryEXAM', 'studyHOURS', 'Python', 'DB']

print("\nCategorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
```

```python
#2. Detect inconsistent categories
# Step 1: Show unique values before cleaning
print("Unique values in gender (before):", df['gender'].unique())
print("Unique values in country (before):", df['country'].unique())
print("Unique values in prevEducation (before):", df['prevEducation'].unique())

# Step 2: Fix inconsistent categories

# Clean gender
df['gender'] = df['gender'].str.strip().str.capitalize()
df['gender'] = df['gender'].replace({
    'M': 'Male',
    'F': 'Female',
    'male': 'Male',
    'female': 'Female'
})

# Clean country
df['country'] = df['country'].str.strip().str.title()

# Clean prevEducation
df['prevEducation'] = df['prevEducation'].str.strip().str.title()
df['prevEducation'] = df['prevEducation'].replace({
    'Barrrchelors': 'Bachelors',
    'Bachlors': 'Bachelors',
    'Highschool': 'High School'
})

# Step 3: Show unique values after cleaning
print("\nUnique values in gender (after):", df['gender'].unique())
print("Unique values in country (after):", df['country'].unique())
print("Unique values in prevEducation (after):", df['prevEducation'].unique())
```

```python
#3. Handle duplicates
# Step 1: Check number of duplicate rows
duplicates_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicates_count)

# Step 2: Drop duplicates
df = df.drop_duplicates()

# Step 3: Confirm duplicates removed
duplicates_after = df.duplicated().sum()
print("Number of duplicate rows after cleaning:", duplicates_after)
```

```python
# Part 2 - Missing Data

# Step 1: Identify missing values in each column
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)

# Step 2: Find which columns are most affected
most_affected = missing_values.sort_values(ascending=False)
print("\nMost affected columns by missing values:\n", most_affected[most_affected > 0])
```

```python
#2. Impute missing values
# Option 1: Numerical imputation (mean / median)
# Fill missing values in 'Python' with mean
df_mean = df.copy()
df_mean['Python'] = df_mean['Python'].fillna(df_mean['Python'].mean())

# Fill missing values in 'Python' with median
df_median = df.copy()
df_median['Python'] = df_median['Python'].fillna(df_median['Python'].median())

print("After mean imputation (Python):\n", df_mean['Python'])
print("\nAfter median imputation (Python):\n", df_median['Python'])

# Option 2: Categorical imputation (mode)
# Example with 'gender'
df_mode = df.copy()
df_mode['gender'] = df_mode['gender'].fillna(df_mode['gender'].mode()[0])

# Example with 'country'
df_mode['country'] = df_mode['country'].fillna(df_mode['country'].mode()[0])

print("\nAfter mode imputation (gender):\n", df_mode['gender'])
print("\nAfter mode imputation (country):\n", df_mode['country'])
```

```python
# Part 3 - Outliers

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Detect outliers using Boxplots
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(x=df['studyHOURS'])
plt.title("Outliers in studyHOURS")

plt.subplot(1,3,2)
sns.boxplot(x=df['Python'])
plt.title("Outliers in Python")

plt.subplot(1,3,3)
sns.boxplot(x=df['DB'])
plt.title("Outliers in DB")

plt.show()


# 2. Detect outliers using Summary Statistics
print("StudyHOURS stats:\n", df['studyHOURS'].describe(), "\n")
print("Python stats:\n", df['Python'].describe(), "\n")
print("DB stats:\n", df['DB'].describe(), "\n")
```

```python
# Option 2: Apply IQR method for studyHOURS, Python, DB
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Apply on numerical columns
df_iqr = df.copy()
for col in ['studyHOURS', 'Python', 'DB']:
    df_iqr = remove_outliers_iqr(df_iqr, col)

print("Shape before removing:", df.shape)
print("Shape after IQR filtering:", df_iqr.shape)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("bi.csv", encoding="utf-8")

# Normalize column names (lowercase)
df.columns = df.columns.str.lower()

# Standardize gender values
df["gender"] = df["gender"].replace({
    "M": "Male", "F": "Female"
})

# Standardize previous_education values
df["previous_education"] = df["previous_education"].replace({
    "diploma": "Diploma",
    "DIPLOMA": "Diploma",
    "Diplomaaa": "Diploma",
    "Barrrchelors": "Bachelors"
})

# Handle missing values in Python (if any) by filling with mean
if df["python"].isnull().sum() > 0:
    df["python"].fillna(df["python"].mean(), inplace=True)

# Detect and remove outliers in Age using IQR
q1 = df["age"].quantile(0.25)
q3 = df["age"].quantile(0.75)
iqr = q3 - q1
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr

df = df[(df["age"] >= lower) & (df["age"] <= upper)]

# Save cleaned dataset
df.to_csv("students.csv", index=False)

print("Cleaned dataset saved as students.csv")
```
